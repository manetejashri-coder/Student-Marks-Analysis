```python
import pandas as pd
```

```python
data = pd.read_csv("students.csv") #Load dataset
print(data)

print("Average Marks:",data["Marks"].mean())
print("Highest Marks:",data["Marks"].max())
print("Lowest Marks:",data["Marks"].min())

data["Result"] = data["Marks"].apply(lambda x:"Pass" if x>=40 else"Fail")
print(data)

import matplotlib.pyplot as plt

plt.bar(data["Name"],data["Marks"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

print(data["Result"].value_counts())

top_student = data[data["Marks"] == data["Marks"].max()]
print(top_student)

fail_students = data[data["Result"] == "Fail"]
print(fail_students)


```

Student Marks Analysis Project

Steps included:
- Loaded dataset from csv file
- Calculated average, highest, and lowest marks
- Classified students as Pass or Fail
 - Created a bar chart for visualization
